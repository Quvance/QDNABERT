```python
def list_files(dir):
    
    r=[]
    subdirs=[x[0] for x in os.walk(dir)]
    for subdir in subdirs:
        files=os.walk(subdir).__next__()[2]
        if (len(files)>0):
            for file in files:
                r.append(os.path.join(subdir,file))
    return r
                
                
dirpath="/Users/brah/Documents/Quantum/Quantum BERT/code/QDNABERT-main/results/DNABERT2/"

list=list_files(dirpath)
print(list_files)

for pathtofile in list:
    if (os.path.isfile(pathtofile)):
        
        strPath = os.path.realpath(pathtofile)
        nmFolders = strPath.split( os.path.sep )
       
        print( f"Class  :{nmFolders[-4]}" )
        print( f"Subclass:{nmFolders[-3]}" )
        try:
            with open (pathtofile) as user_file:
                dict=json.load(user_file)
                for key in dict:
                    print(f'{key:30} -> {dict[key]:20}')
        except:
            pass
            
        
    


        
```
